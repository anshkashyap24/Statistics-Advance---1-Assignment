13. What is the Central Limit Theorem (CLT), and why is it critical in statistics ?

 - Allows Statistical Inference
It lets us make predictions about sample means using the normal distribution, even if the population isn't normal.

 - Justifies Hypothesis Testing and Confidence Intervals

 - Works for Many Real-World Problems

n.

 - Foundation of Modern Statistics
It underlies techniques like regression, ANOVA, and much more.

14.  How does the Central Limit Theorem relate to the normal distribution ?
 - The CLT says that the distribution of the sample means (or sums) will tend toward a normal distribution as the sample size increases, regardless of the shape of the population distribution.

So, even if your data is not normal, the sampling distribution of the mean will be approximately normal ‚Äî and that normal shape comes directly from the normal distribution.

21.  What are the assumptions for applying the Central Limit Theorem ?
 - 1. Independence of Observations
The data points in the sample should be independent of one another.

  - This usually means using random sampling or ensuring that the sample doesn‚Äôt influence itself (no repeated measurements on the same subject unless modeled correctly).
 - 2. Sample Size Is Sufficiently Large
For non-normal populations, a larger sample size is required for the CLT to hold.

  - Rule of thumb: n ‚â• 30 for skewed or unknown distributions.

  - If the population is already normal, smaller samples (even n < 30) are usually fine.

  - 3. Finite Variance and Mean
The population from which you're sampling must have a finite mean and finite standard deviation.

 - Extreme outliers or infinite variance (as in some theoretical distributions) can violate this condition.

 - 4. Random Sampling
 - The sample should be drawn randomly from the population to avoid bias and ensure representativeness.

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution .
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    float: The probability of observing x
    """
    if x not in [0, 1]:
        return 0  # Bernoulli only defined for 0 and 1
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_scores(data):
    """
    Compute Z-scores for a given dataset.

    Parameters:
    data (array-like): Input dataset

    Returns:
    np.ndarray: Z-score normalized dataset
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = (data - mean) / std_dev
    return z_scores

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

z_scores = calculate_z_scores(data)

x = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x, 0, 1)

plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', label="Z-score Distribution")

plt.plot(x, pdf_values, 'r-', linewidth=2, label="Standard Normal PDF")

plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Z-score Distribution and Standard Normal Curve")
plt.legend()
plt.grid()

plt.show()
```

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    """
    Compute the Z-score for a given data point in a dataset.

    Parameters:
    - data_point (float): The value for which the Z-score is calculated.
    - data (array-like): The dataset.

    Returns:
    - float: Z-score of the data point.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_score = (data_point - mean) / std_dev
    return z_score

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
data_point = 60

z_score = calculate_z_score(data_point, data)
print(f"Z-score for data point {data_point}: {z_score:.2f}")

x = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x, 0, 1)

plt.plot(x, pdf_values, 'r-', linewidth=2, label="Standard Normal Distribution (N(0,1))")

plt.axvline(z_score, color='blue', linestyle='--', label=f"Z-score = {z_score:.2f}")
plt.fill_between(x, pdf_values, where=(x < z_score), color='blue', alpha=0.2)

plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Z-score and Standard Normal Distribution")
plt.legend()
plt.grid()

plt.show()
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test(sample, pop_mean, alpha=0.05, alternative="two-tailed"):
    """
    Performs a Z-test for the population mean based on the sample.

    Parameters:
    - sample: List or array of sample data
    - pop_mean: Hypothesized population mean (H0)
    - alpha: Significance level (default=0.05)
    - alternative: "two-tailed", "left-tailed", or "right-tailed"

    Returns:
    - Z-score and p-value
    - Decision on whether to reject H0
    """
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)

    se = sample_std / np.sqrt(sample_size)
    z_score = (sample_mean - pop_mean) / se

    if alternative == "two-tailed":
        z_critical = norm.ppf(1 - alpha/2)
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
        reject_h0 = abs(z_score) > z_critical
    elif alternative == "left-tailed":
        z_critical = norm.ppf(alpha)
        p_value = norm.cdf(z_score)
        reject_h0 = z_score < z_critical
    elif alternative == "right-tailed":
        z_critical = norm.ppf(1 - alpha)
        p_value = 1 - norm.cdf(z_score)
        reject_h0 = z_score > z_critical

    print(f"Sample Mean: {sample_mean:.2f}, Population Mean (H0): {pop_mean}")
    print(f"Z-score: {z_score:.2f}, Critical Z: {z_critical:.2f}, p-value: {p_value:.4f}")
    print("Reject H0" if reject_h0 else "Fail to Reject H0")

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.plot(x, y, 'r-', linewidth=2, label="Standard Normal Distribution")
    plt.axvline(z_score, color='blue', linestyle='--', label=f"Z-score = {z_score:.2f}")

    if alternative == "two-tailed":
        plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label="Rejection Region")
    elif alternative == "left-tailed":
        plt.fill_between(x, y, where=(x < z_critical), color='red', alpha=0.3, label="Rejection Region")
    elif alternative == "right-tailed":
        plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3, label="Rejection Region")

    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Hypothesis Testing using Z-Statistics")
    plt.legend()
    plt.grid()
    plt.show()

np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)
population_mean_h0 = 50

z_test(sample_data, population_mean_h0, alpha=0.05, alternative="two-tailed")
```

```python
# Create a confidence interval for a dataset using Python and interpret the result?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Computes the confidence interval for a dataset.

    Parameters:
    - data: List or array of sample data.
    - confidence: Confidence level (default=95%).

    Returns:
    - Tuple: (lower bound, upper bound) of the confidence interval.
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    se = sample_std / np.sqrt(n)

    alpha = 1 - confidence
    z_critical = norm.ppf(1 - alpha/2)

    margin_of_error = z_critical * se
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

conf_level = 0.95
lower, upper, mean, margin = confidence_interval(sample_data, conf_level)

print(f"Sample Mean: {mean:.2f}")
print(f"{conf_level*100:.0f}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Margin of Error: ¬±{margin:.2f}")

x = np.linspace(mean - 4*margin, mean + 4*margin, 1000)
y = norm.pdf(x, loc=mean, scale=margin)

plt.plot(x, y, 'r-', linewidth=2, label="Sample Mean Distribution")
plt.axvline(lower, color='blue', linestyle='--', label=f"Lower Bound ({lower:.2f})")
plt.axvline(upper, color='blue', linestyle='--', label=f"Upper Bound ({upper:.2f})")
plt.axvline(mean, color='black', linestyle='-', label=f"Sample Mean ({mean:.2f})")
plt.fill_between(x, y, where=(x >= lower) & (x <= upper), color='blue', alpha=0.2)

plt.xlabel("Sample Mean")
plt.ylabel("Probability Density")
plt.title(f"{conf_level*100:.0f}% Confidence Interval for Population Mean")
plt.legend()
plt.grid()

plt.show()
```

```python
# Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std=1):
    """
    Plots the probability density function (PDF) of a normal distribution.

    Parameters:
    - mean: Mean (Œº) of the normal distribution (default=0).
    - std: Standard deviation (œÉ) of the normal distribution (default=1).
    """
    x = np.linspace(mean - 4*std, mean + 4*std, 1000)

    y = norm.pdf(x, mean, std)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, 'r-', linewidth=2, label=f'Normal(Œº={mean}, œÉ={std})')

    plt.axvline(mean, color='black', linestyle='--', label=f"Mean (Œº={mean})")
    plt.axvline(mean - std, color='blue', linestyle='--', alpha=0.7, label=f"1 Std Dev (Œº-œÉ)")
    plt.axvline(mean + std, color='blue', linestyle='--', alpha=0.7, label=f"1 Std Dev (Œº+œÉ)")

    plt.xlabel("X-values")
    plt.ylabel("Probability Density")
    plt.title("Normal Distribution PDF")
    plt.legend()
    plt.grid()

    plt.show()

plot_normal_pdf(mean=0, std=1)
```

```python
# Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam=5, max_k=15):
    """
    Plots the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
    - lam: Mean number of events (Œª) per interval.
    - max_k: Maximum number of occurrences to display in the plot.
    """

    k_values = np.arange(0, max_k + 1)

    cdf_values = poisson.cdf(k_values, lam)

    plt.figure(figsize=(8, 5))
    plt.step(k_values, cdf_values, where='mid', color='red', linewidth=2, label=f'Poisson(Œª={lam}) CDF')

    plt.xlabel("Number of Events (k)")
    plt.ylabel("Cumulative Probability")
    plt.title(f"Poisson Distribution CDF (Œª={lam})")
    plt.xticks(k_values)
    plt.yticks(np.arange(0, 1.1, 0.1))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()

    plt.show()

plot_poisson_cdf(lam=5, max_k=15)
```

```python
# Simulate a random variable using a continuous uniform distribution and calculate its expected value?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def simulate_uniform_distribution(a=0, b=10, size=1000):
    """
    Simulates a continuous uniform distribution and calculates its expected value.

    Parameters:
    - a: Lower bound of the distribution.
    - b: Upper bound of the distribution.
    - size: Number of random samples to generate.

    Returns:
    - Expected value of the uniform distribution.
    """

    data = np.random.uniform(a, b, size)

    expected_value = (a + b) / 2

    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')

    x = np.linspace(a, b, 1000)
    y = uniform.pdf(x, a, b-a)
    plt.plot(x, y, 'r-', lw=2, label=f'Uniform({a}, {b}) PDF')

    plt.axvline(expected_value, color='black', linestyle='--', label=f'Expected Value (E[X] = {expected_value})')
    plt.xlabel("X-values")
    plt.ylabel("Density")
    plt.title("Continuous Uniform Distribution")
    plt.legend()
    plt.grid()

    plt.show()

    return expected_value

expected_val = simulate_uniform_distribution(a=0, b=10, size=1000)

print(f"Expected Value (Theoretical): {expected_val:.2f}")
```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference?

import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(size=1000, mean1=50, std1=10, mean2=50, std2=20):
    """
    Generates two datasets with different standard deviations, compares them, and visualizes the difference.

    Parameters:
    - size: Number of samples in each dataset.
    - mean1: Mean of the first dataset.
    - std1: Standard deviation of the first dataset.
    - mean2: Mean of the second dataset.
    - std2: Standard deviation of the second dataset.
    """

    data1 = np.random.normal(loc=mean1, scale=std1, size=size)
    data2 = np.random.normal(loc=mean2, scale=std2, size=size)

    std_dev1 = np.std(data1, ddof=1)
    std_dev2 = np.std(data2, ddof=1)

    print(f"Dataset 1 - Standard Deviation: {std_dev1:.2f}")
    print(f"Dataset 2 - Standard Deviation: {std_dev2:.2f}")

    plt.figure(figsize=(10, 5))
    plt.hist(data1, bins=30, alpha=0.6, color='blue', edgecolor='black', label=f'Dataset 1 (œÉ={std_dev1:.2f})')
    plt.hist(data2, bins=30, alpha=0.6, color='red', edgecolor='black', label=f'Dataset 2 (œÉ={std_dev2:.2f})')

    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title("Comparison of Standard Deviations")
    plt.legend()
    plt.grid()

    plt.show()

compare_standard_deviations(size=1000, mean1=50, std1=10, mean2=50, std2=20)
```

```python
# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?

import numpy as np
import matplotlib.pyplot as plt

def calculate_range_and_iqr(mean=50, std=15, size=1000):
    """
    Generates a normally distributed dataset, calculates its range and interquartile range (IQR),
    and visualizes it using a box plot.

    Parameters:
    - mean: Mean of the normal distribution.
    - std: Standard deviation of the normal distribution.
    - size: Number of samples.

    Returns:
    - Range and IQR of the dataset.
    """
    data = np.random.normal(loc=mean, scale=std, size=size)

    data_range = np.max(data) - np.min(data)

    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    iqr = Q3 - Q1

    print(f"Range: {data_range:.2f}")
    print(f"Interquartile Range (IQR): {iqr:.2f}")

    plt.figure(figsize=(8, 5))
    plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))

    plt.xlabel("Values")
    plt.title("Box Plot of Normally Distributed Data")
    plt.grid()

    plt.show()

    return data_range, iqr

range_value, iqr_value = calculate_range_and_iqr(mean=50, std=15, size=1000)
```

```python
#  Implement Z-score normalization on a dataset and visualize its transformation?

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def z_score_normalization(mean=50, std=15, size=1000):
    """
    Generates a normally distributed dataset, applies Z-score normalization, and visualizes
    the transformation.

    Parameters:
    - mean: Mean of the original normal distribution.
    - std: Standard deviation of the original normal distribution.
    - size: Number of samples.

    Returns:
    - Original and normalized datasets.
    """

    data = np.random.normal(loc=mean, scale=std, size=size)

    data_mean = np.mean(data)
    data_std = np.std(data, ddof=1)

    normalized_data = (data - data_mean) / data_std

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(data, bins=30, kde=True, color='blue', edgecolor='black')
    plt.axvline(data_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {data_mean:.2f}')
    plt.title("Original Data Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid()

    plt.subplot(1, 2, 2)
    sns.histplot(normalized_data, bins=30, kde=True, color='green', edgecolor='black')
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean: 0")
    plt.axvline(1, color='black', linestyle='dashed', linewidth=1, label="Std Dev: 1")
    plt.axvline(-1, color='black', linestyle='dashed', linewidth=1)
    plt.title("Z-score Normalized Data")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

    return data, normalized_data

original_data, normalized_data = z_score_normalization(mean=50, std=15, size=1000)
```
